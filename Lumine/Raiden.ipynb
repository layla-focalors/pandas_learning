```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('../data/billboard.csv')
df
```

```python
df.columns
```

```python
df = pd.melt(df, id_vars=['year','artist','track','time','date.entered'], var_name="week",value_name="rating")
df
```

```python
ax = sns.displot(x='rating', y='week',data=df)
ax.tight_layout()
```

```python
sns.scatterplot(x='rating', y='week', data=df)
```

```python
sns.lmplot(x='rating', y='rating', data=df, col='artist', col_wrap=4)
```
