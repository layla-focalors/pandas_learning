```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data/gapminder.tsv', sep='\t')
df.head()
```

```python
ax = sns.FacetGrid(df, col='continent')
ax.map(sns.boxplot, 'year', 'pop')
```

```python
ax = sns.FacetGrid(df, col='continent')
ax.map(sns.barplot, 'year', 'pop', errorbar=None)
```

```python
ax = sns.FacetGrid(df, col='continent', hue='country')
ax.map(sns.barplot, 'year', 'pop', errorbar=None)
```

```python
ax = sns.FacetGrid(df, col='country', col_wrap=4)
ax.map(sns.boxplot, 'year', 'pop')
```

```python
ax = sns.FacetGrid(df, col='country', col_wrap=4)
ax.map(sns.lineplot, 'year', 'pop')
```

```python
ax = sns.FacetGrid(df, col='country', col_wrap=4)
ax.map(sns.lineplot, 'gdpPercap', 'pop')
```

```python
ax = sns.FacetGrid(df, col='country', col_wrap=4)
ax.map(sns.scatterplot, 'gdpPercap', 'pop')
```
