```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data/weather.csv')
df.head()
```

```python
melter = pd.melt(df, id_vars=['id', 'year', 'month', 'element'])
melter
```

```python
ax = melter.pivot_table(index=['id', 'year', 'variable', 'month'], columns='element')
ax
```

```python
an = ax.reset_index()
an
```

```python
an.head(1)
```

```python
an = pd.melt(an, id_vars=['id','year', 'month', 'variable'])
an
```

```python
an.columns = ['id', 'year', 'month', 'values', 'None', 'element', 'value']
an = an.drop(['None'], axis=1)
an
```

```python
an.columns = ['id', 'year', 'month', 'value', 'value', 'element', 'value']
```
