```python
import pandas as pd
df = pd.read_csv('../data/weather.csv')
df.head()
```

```python
melted = pd.melt(df, id_vars=['id','year','month', 'element'], var_name="day", value_name="temp")
melted
```

```python
df = pd.pivot_table(melted, index=['id', 'year', 'month','day'], columns='element')
df
```
