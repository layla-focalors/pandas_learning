```python
# 복구 시 remelt 절차 필수
remelt = io.melt(id_vars=['id','year','month', 'variable'])
remelt
```

```python
remelt.columns = ['id', 'year', 'month', 'day', 'None', 'element', 'value']
remelt = remelt.drop(['None'], axis=1)
```

```python
df.head(1)
```

```python
rm = remelt.pivot(columns=['day'], values='value')
rm
```

```python
melter = pd.concat([remelt, rm])
melter
```

```python
import pandas as pd

# DataFrame 생성
df = pd.DataFrame({
    'd1': [1, 2, 3],
    'd10': [4, 5, 6],
    'd2': [7, 8, 9]
})

# 열 이름을 숫자 순서대로 정렬
columns_sorted = sorted(df.columns, key=lambda x: int(x[1:]))

# 열을 숫자 순서대로 정렬
df_sorted = df.reindex(columns=columns_sorted)

print(df_sorted)
```
