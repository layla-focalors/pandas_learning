```python
import pandas as pd
```

```python
# seaborn의 각 차트들 그려보기
import seaborn as sns
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
ax = sns.barplot(x='day', y='tip', data=df, errorbar=None)
ax.set_title("barplot for day & tips")
```

```python
sns.countplot(x='day', data=df)
# 빈도를 세는 countplot
```

```python
# 단변량
sns.distplot(df['total_bill'], kde=False)
```

```python
# 단변량
sns.distplot(df['total_bill'], kde=True, rug=True)
```

```python
# 단변량
sns.distplot(df['total_bill'], hist=False)
```

```python
sns.displot(df['total_bill'])
```

```python
sns.histplot(x='total_bill', y='tip', data=df)
```

```python
sns.histplot(df[['total_bill', 'tip']])
```

```python
ax = sns.kdeplot(x='total_bill', y="tip", data=df, fill=True)
ax.set_title("kde plot")
```

```python
sns.boxplot(x='day', y='tip', data=df, hue='sex')
```

```python
sns.pairplot(df)
```

```python
sns.jointplot(x='total_bill', y='tip', data=df)
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind="hex", color='pink')
```

```python
ax = sns.FacetGrid(df, col='day')
ax.map(sns.scatterplot, "total_bill", "tip")
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
```

```python
sns.lmplot(x='total_bill', y='tip', data=df)
```

```python
sns.regplot(x='total_bill', y='tip', data=df)
```
