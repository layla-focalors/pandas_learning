```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data/country_timeseries.csv')
df.head()
```

```python
data = pd.melt(df, id_vars=['Date', 'Day'], var_name="case", value_name="count")
data
```

```python
table = data['case'].str.split("_", expand=True)
table.columns = ['Status', 'Country']
table
```

```python
df = pd.concat([data, table])
df
```

```python
table2 = df['case'].str.split("_")
table2
```

```python
df['status'] = table2.str.get(0)
df['country'] = table2.str.get(1)
```
