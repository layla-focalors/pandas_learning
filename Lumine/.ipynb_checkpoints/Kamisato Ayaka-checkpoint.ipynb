```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
axes = sns.jointplot(x="total_bill", y="tip", data=df, kind='hex', color='pink')
axes.set_axis_labels(xlabel="total bills", ylabel="tips")
axes.fig.suptitle("Snow on the Hearth")
plt.subplots_adjust(top=0.95)
```
