```python
melted = pd.melt(df, id_vars=['character_name', 'rarity', 'region', 'vision', 'weapon_type', 'model',
       'constellation', 'birthday', 'special_dish', 'affiliation', 'voice_eng',
       'voice_cn', 'voice_jp', 'voice_kr', 'release_date', 'ascension',
       'ascension_specialty', 'ascension_talent', 'ascension_boss',
       'talent_book', 'talent_weekly'], var_name="stats", value_name="data") 
melted
```

```python
df.shape
```

```python
melted.shape
```

```python
melted.columns
```

```python
melted['id'] = range(len(melted))
```

```python
melted.columns
```

```python
df = pd.pivot_table(melted, index=['id','character_name','rarity', 'region', 'vision'], columns='stats') 
df
```

```python
df.interpolate()
```

```python
df = df.reset_index()
```
