```python
df = pd.read_csv('../data/billboard.csv')
df.head()
```

```python
melted = pd.melt(df, id_vars=['year', 'artist', 'track', 'time', 'date.entered'], var_name="week", value_name="rating")
melted
```

```python
melted_subset = melted[['year', 'artist', 'track', 'time', 'date.entered']]
melted_subset.drop_duplicates()
```

```python
melted_subset['id'] = range(len(melted_subset))
```

```python
melted_subset
```

```python
df = melted.merge(melted_subset, on=['year', 'artist', 'track', 'time', 'date.entered'])
df
```

```python
df.columns
```

```python
ax = sns.FacetGrid(df, col='track', col_wrap=4)
ax.map(sns.barplot, "week", "rating")
```
