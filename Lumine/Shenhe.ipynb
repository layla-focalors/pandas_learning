```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data/weather.csv')
df.head()
```

```python
melter = pd.melt(df, id_vars=['id', 'year', 'month', 'element'], var_name="day", value_name="temp")
melter
```

```python
df = melter.pivot_table(index=['id', 'year','month','day'], columns=['element'])
df
```

```python
df = df.reset_index()
df
```

```python
df = pd.melt(df, id_vars=['index','id', 'year', 'month', 'day'])
df
```

```python
df.pivot(index=['id'], columns=[ 'year', 'month'], values='value')
```
