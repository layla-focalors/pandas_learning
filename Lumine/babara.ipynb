```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.histplot(x='day', y='tip', data=df)
```

```python
# 팁이 목요일과 토요일, 일요일에 몰리는 것으로 보아, 대부분이 목, 토요일에 방문할 것이다.
```

```python
sns.histplot(x='time', y='day', data=df)
```

```python
# 대부분의 손님이 저녁, 토, 일에 방문한다.
```

```python
axes = sns.FacetGrid(df, hue='sex', row='time', col='day')
axes.map(sns.scatterplot, "total_bill", "tip")
```

```python
# 산포도를 뿌려 보면 위의 그래프와 같이 주말 오후, 목요일 오전에 사람들이 몰리는 경향이 있음
```

```python
sns.histplot(x='size', y='day', data=df)
```

```python
# 이때 인원 대부분은 2인 손님이고 목요일과 토요일에 많이 방문.
# 이에, 인원이 많으면 팁을 많이 줄 것이고 금액이 클 것이다. 라는 가설 설정
```

```python
sns.lmplot
```

```python
sns.lmplot(col='size', x='total_bill', y='tip', data=df, fit_reg=False, hue='sex')
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='size', data=df)
```

```python
sns.histplot(x='time', y='day', data=df)
```

```python
sns.histplot(x='sex', y='time', data=df)
```

```python
sns.histplot(x='sex', y='day', data=df)
```

```python
sns.histplot(x='sex', y='total_bill', data=df)
```

```python
# 남자의 금액 시작이 조금 더 높은 상태임을 알 수 있다!
```

```python
sns.histplot(x='sex', y='tip', data=df)
```

```python
# 위의 금액 시작과 아래의 팁을 대조하면, 대부분의 금액은 10~20달러선이고, 팁도 2~4달러 선에 머무는 경우가 많음을 알 수 있음!
```
