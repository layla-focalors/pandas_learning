```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset("tips")
```

```python
sns.FacetGrid
```

```python
sns.boxplot(x="time", y="tip", data=df,hue="sex")
# 위의 박스플롯에 따라서 tip은 저녁 시간에 많이 평균값이 크고, 이상치가 있는 것으로 보아
# 저녁 시간에 인기 있는 식당이라는 곳을 알 수 있음 그에 따라서
```

```python
sns.barplot(x="day", y="size", data=df, errorbar=None)
```

```python
# 시간대별 손님의 방문 예측
# 위의 결과에 따른 가설을 세우면,
# 가설 : 팁이 많이 제공되는 저녁에 대부분의 손님들이 위치할 것이다.
ax = sns.FacetGrid(df, hue="sex", col="day", row="time")
ax.map(sns.scatterplot, "total_bill", "tip")
```

### 결과 해석
일요일과 토요일에는 Dinner시간에만 손님이 몰리고,   
금요일은 골고루 퍼져 있고,  
목요일은 오전에만 대부분의 손님이 방문  

```python
# 이에 따라 추정하자면, 직장인 등이 주 고객인 것 같으며, 토요일과 일요일 오후에 남/여 의 방문이
# 잘 나타난다고 해석할 수 있다.

# 그 외에 추세를 다시 보자면,
# 목요일 오전에는 여성 손님이 다소 많은 것으로 보여진다.
```

```python
# 연계가설 : 이 식당에서는 다소 가벼운 메뉴와 무거운 ( 주말 )등에 회식 등에 사용되는
# 메뉴를 동시에 갖추고 있을 것으로 사료된다. 
```

```python
sns.kdeplot(x="total_bill", y="tip", fill=True, data=df)
```

```python
# 또한 kde 플롯을 통해 분포 위치를 추정하면, 대부분의 금액과 팅비 10-20 달러 선에 묶여 있음을 확인할 수 있다
```

```python
sns.displot(df['total_bill'])
```

```python
sns.lmplot(x="total_bill", y="tip", data=df, hue="sex", fit_reg=False)
```

```python
sns.boxplot(x="day", y="tip", data=df)
```

```python
# 새로운 가설. 어느 날짜에 단체 손님들이 많이 방문할까? ( 4인 이상 )
```

```python
sns.histplot(x="size", y="day", data=df)
```

```python
# 위의 결과에 따르면, 2인 손님 비율이 전체 테이블 수에서 압도적이다. 이에 반해,
# 4인 , 6인 손님들은 일요일과 목요일 혹은 토요일에 방문율이 큰 것을 알 수 있으며
# 전체적으로 토요일이 제일 유망하다는 것을 알 수 있다.
# 이에 컨설팅을 하자면, 2인 테이블을 토요일에 늘리는 것이 좋은 것 같다.
```

```python
# 또한 scatter 플롯을 그렸을 때, 대부분의 위치가 낮은 금액에 위치하는 것으로 보아
# 목요일 ~ 금요일은 낮은 가격으로 판매하는 작은 레스토랑, 
# 그러나 토/일은 팁이 전체적으로 높고 그 중에 가장 높은 날은 토요일인 것으로 추정된다.
```

```python
# 그러면 어떤 성별이 가장 팁을 많이 줄까?
```

```python
sns.histplot(x='sex', y='tip', data=df)
```

```python
# 전체적인 팁의 분포는 남성에서 높게 나타나며, 대부분 2~4달러 사이의 팁이 제공된다.
# 이에 짙은 색이 나타나는 것으로 보아 남성측이 더 많은 팁을 제공하고 있음을 알 수 있다.
```

```python
# 그렇다면, 남성과 여성은 어떤 시간대에 많이 방문할까?
```

```python
sns.histplot(x="sex", y="time", data=df)
```

```python
# 위 히스토그램에 따르면, 대부분(짙은) 남성 측이 저녁에 가장 많이 방문하고, 여성 측도 점심때에 비해서
# 저녁에 많이 방문하는 모습을 볼 수 있다.
```

```python
df[df['sex'] == 'Male'].shape[0]
```

```python
df[df['sex'] == 'Female'].shape[0]
```

```python
# 또한 이 결과를 뒷받침하기 위해서 각 성별로 데이터의 수를 보면, 남성이 157, 여성이 87개의 관측값으로
# 남성의 데이터 수 자체가 많고, 위 결과를 증명한다
# 즉, 조언을 하나 더 하자면, 남성 측이 선호하는 메뉴를 추가하거나, 여성 측이 선호하는 메뉴를 추가하여
# 성별의 비율을 맞추는 편이 좋아보인다.
```

```python
male = df[df['sex'] == 'Male']
female = df[df['sex'] == 'Female']
```

```python
# 그러면, 어느 쪽의 지출 금액이 큰가?
```

```python
sns.lmplot(col='day', x='total_bill', y='tip', data=male, fit_reg=False)
```

```python
male['total_bill'].mean()
```

```python
male['total_bill'].median()
```

```python
sns.lmplot(col='day', x='total_bill', y='tip', data=female, fit_reg=False)
```

```python
female['total_bill'].mean()
```

```python
female['total_bill'].median()
```

```python
# 위 차트를 보면, 일자별 각 성별의 차이를 나타낸다, 평균 금액은 남성이 크다. 중앙값도 남성이 높은 것으로,
# 나타남에 따라 각 차트별로 보면 남성 쪽이 더 높을 것으로 보인다.
```

```python
sns.boxplot(x="sex", y="total_bill", data=df)
```

```python
sns.boxplot(x="sex", y="tip", data=df)
```
