```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
ax = sns.FacetGrid(df, col='day', row='time')
ax.map(sns.boxplot, 'total_bill', 'tip')
```

```python
ax = sns.FacetGrid(df, col='day', row='time')
ax.map(sns.histplot, 'total_bill', 'tip')
```

```python
ax = sns.FacetGrid(df, col='day', row='time', hue='size')
ax.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
ax = sns.FacetGrid(df, col='day', row='time', hue='size')
ax.map(sns.barplot, 'total_bill', 'tip')
```

```python
ax = sns.FacetGrid(df, col='day', row='time', hue='sex')
ax.map(sns.scatterplot, 'smoker', 'tip')
```
