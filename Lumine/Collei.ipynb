```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data/weather.csv')
df.head(3)
```

```python
melted = pd.melt(df, id_vars=['id', 'year', 'month', 'element'])
melted
```

```python
vm = melted.pivot_table(index=['id','year','month', 'variable'], columns=['element'])
vm
```

```python
import os
os.mkdir('./file')
```

```python
# 인덱스 복구하기
io = vm.reset_index()
io
```

```python
io.columns
```

```python
# 복구 시 remelt 절차 필수
remelt = io.melt(id_vars=['id','year','month', 'variable'])
remelt
```

```python
remelt.columns = ['id', 'year', 'month', 'day', 'None', 'element', 'value']
remelt = remelt.drop(['None'], axis=1)
```

```python
df.head(1)
```

```python
rm = remelt.pivot(columns=['day'], values='value')
rm
```

```python
df.head(2)
```

```python
melter = pd.concat([remelt, rm])
melter
```

```python
melter = melter.drop(['day', 'value'], axis=1)
```
