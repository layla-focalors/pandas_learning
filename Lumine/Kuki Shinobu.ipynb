```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.regplot(x='total_bill', y='tip', data=df, fit_reg=False)
```

```python
sns.regplot(x='total_bill', y='tip', data=df)
```

```python
sns.regplot
```

```python
sns.lmplot(x='total_bill', y='tip', data=df, hue='day')
```

```python
sns.regplot(x='total_bill', y='tip', data=df)
```

```python
ax = sns.FacetGrid(df, col='day', hue='sex', row='time')
ax.map(sns.scatterplot, "total_bill", "tip")
```

```python
ax = sns.FacetGrid(df, col='day', hue='sex', row='time')
ax.map(sns.barplot, "total_bill", "tip")
```

```python
ax = sns.FacetGrid(df, col='day', hue='sex')
ax.map(sns.barplot, "total_bill", "time", errorbar=None)
```

```python
ax = sns.FacetGrid(df, col='day', row='time')
ax.map(sns.boxplot, "sex", "total_bill")
```
