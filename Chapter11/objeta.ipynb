```python
def mean_x2(X):
    n = len(X)
    sum = 0
    for i in range(n):
        sum += int(X[i])
    return sum / n
#  ! !
```

```python
def my_mean(X):
    n = len(X)
    sum = 0
    
    for i in X:
        sum += i
    
    return sum / n
```

```python
a_my_mean = df.groupby('year').lifeExp.agg(my_mean)
```

```python
a_my_mean = df.groupby('year').lifeExp.agg(mean_x2)
```

```python
a_my_mean
```

```python
def my_mean_diff(x, y):
    n = len(x)
    sum = 0
    for i in x:
        sum += i
    mean = sum / n
    return mean - y
```

```python
global_mean = df.lifeExp.mean()
```

```python
global_mean
```

```python
df.groupby('year').lifeExp.agg(my_mean_diff, y=df.lifeExp.mean())
```

```python
agg_mean_diff = df.groupby('year').lifeExp.agg(my_mean_diff, y=global_mean)
```

```python
agg_mean_diff
```

```python
df.groupby('year').lifeExp.apply(my_mean_diff, y=global_mean)
```
