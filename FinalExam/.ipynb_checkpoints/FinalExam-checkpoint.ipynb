```python
# 사용자 지정 함수를 이용한 연산
import numpy as np
def nulltype(data):
    null_x = data.isnull()
    count_null = null_x.sum()
    return count_null

def fx(data):
    return data.isnull().sum()
```

```python
dsum = df.apply(nulltype)
dsum
```

```python
dsum = df.apply(fx)
dsum
```

```python
def meani(data):
    return fx(data) / data.size
```

```python
# 비율구하기 , 결측치를 구한 다음 data.size로 나누기
df.size
```

```python
def prop_complete(data):
    return 1 - meani(data)
```

```python
def prop_completea(data):
    return (1 - meani(data)) * 100
```

```python
df.apply(prop_complete)
```

```python
df.apply(prop_completea)
```

```python
def modelx(data):
    return ((data.isnull().sum()) / data.size) * 100
```

```python
df = pd.read_csv('../data/gapminder.tsv', sep='\t')
```

```python
df.info()
```

```python
df.groupby("year")['lifeExp'].agg(sum)
```

```python
def meaner(data):
    fx = 0
    for i in data:
        fx += i
    return fx / len(data)
```

```python
df.groupby("year")['lifeExp'].agg(meaner)
```

```python
df.groupby("year")['lifeExp'].mean()
```

```python
df.info()
```

```python
def acx(x, y):
    sum = 0
    for i in x:
        sum += i
    return (sum / len(x)) - y
```

```python
df.groupby("year")['lifeExp'].agg(acx, y=1000)
```
