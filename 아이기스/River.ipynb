```python
def my_mean(x):
    sum = 0
    for i in x:
        sum += i
    return sum / len(x)
```

```python
df = pd.read_csv('../data/gapminder.tsv', sep='\t')
```

```python
df
```

```python
df.groupby('year')['lifeExp'].apply(my_mean)
```

```python
df.groupby('year')['lifeExp'].apply(np.mean)
```

```python
def my_mean2(x, n):
    sum = 0
    for i in x:
        sum += i
    return (sum / len(x)) - n
```

```python
df.groupby('year')['lifeExp'].apply(my_mean2, n = df['lifeExp'].mean())
```

```python
df.groupby('year')['lifeExp'].agg([np.count_nonzero, np.mean, np.std])
```

```python
df.groupby('year').agg({'lifeExp':'mean','pop':'median', 'gdpPercap':'median'})
```

```python
def z_score1(x):
    return (x - x.mean()) / x.std()
```

```python
df.groupby('year')['lifeExp'].transform(z_score1)
```

```python
z_score2 = lambda x : (x - x.mean()) / x.std()
```

```python
df.groupby('year')['lifeExp'].transform(z_score2)
```
