```python
df = pd.read_csv('../../data/billboard.csv')
df
```

```python
df.columns
```

```python
melted = pd.melt(df, id_vars=['year', 'artist', 'track', 'time', 'date.entered'], var_name="week", value_name="ratings")
melted
```

```python
cutter = melted[['year','artist','track','time']]
cutter = cutter.drop_duplicates()
cutter['id'] = range(len(cutter))
```

```python
cutter
```

```python
xc = melted.merge(cutter, on=['year', 'artist','track', 'time'])
xc
```

```python
xc.to_csv('./data_processed.csv')
```
