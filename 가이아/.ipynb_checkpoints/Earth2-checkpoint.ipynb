```python
def get_mean(x):
    sum = 0
    for i in x:
        sum += i
    return sum / len(x)

df.groupby('year')['lifeExp'].transform(get_mean)
```

```python
def get_mean(x):
    sum = 0
    for i in x:
        sum += i
    return sum / len(x)

df.groupby('year')['lifeExp'].agg(get_mean)
```

```python
def get_mean(x):
    sum = 0
    for i in x:
        sum += i
    return sum / len(x)

df.groupby('year')['lifeExp'].apply(get_mean)
```

```python
def get_mean(x, n):
    sum = 0
    for i in x:
        sum += i
    return (sum / len(x)) - n

df.groupby('year')['lifeExp'].apply(get_mean, n = df['lifeExp'].mean())
```

```python
df = pd.read_csv('../data/country_timeseries.csv', parse_dates=[0])
```

```python
df
```

```python
df.index = df['Date']
```

```python
df
```

```python
new_idx = pd.date_range(df.index.min(), df.index.max())
```

```python
df = df.reindex(new_idx)
```

```python
df
```

```python
diff = df.apply(pd.Series.last_valid_index) - df.index.min()
```

```python
diff
```

```python
df = df.reindex(new_idx)
```

```python
df
```

```python
dt_lec = {}
for i, v in enumerate(df):
    dt_lec[v] = df[v].shift(diff[i].days)
```

```python
dt_lec
```

```python
df = pd.DataFrame(dt_lec)
```

```python
df
```

```python
df.index = df['Day']
```

```python
df = df.drop(['Date','Day'], axis = 1)
```

```python
df
```

```python
fig = df.plot()
fig.legend(fontsize = 7)
```

```python
fig = sns.lineplot(data = df)
fig.legend(prop={'size':7})
```
