```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
df['missing'] = df.apply(getnan, axis = 1)
```

```python
df
```

```python
def getNAN(x):
    return pd.isnull(x).sum()
```

```python
df['missing2'] = df.apply(getNAN, axis = 1)
```

```python
df
```

```python
getnanp = lambda x : (pd.isnull(x).sum()) / len(x)
```

```python
def getnanp2(x):
    return (pd.isnull(x).sum()) / x.size
```

```python
df.apply(getnanp)
```

```python
df.apply(getnanp2)
```

```python
get_non_nan = lambda x: 1- (pd.isnull(x).sum() / x.size)
```

```python
def get_non_nan2(x):
    return (1 - (pd.isnull(x).sum() / x.size))
```

```python
df.apply(get_non_nan)
```

```python
df.apply(get_non_nan2)
```
