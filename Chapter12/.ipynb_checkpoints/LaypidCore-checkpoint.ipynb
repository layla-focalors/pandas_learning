```python
# 1
df = pd.read_csv('../data/country_timeseries.csv', parse_dates=[0])
```

```python
df.head(10).iloc[:, :5]
```

```python
df.tail(10).iloc[:, :5]
```

```python
df.index = df['Date']
```

```python
df
```

```python
new_idx = pd.date_range(df.index.min(), df.index.max())
```

```python
new_idx
```

```python
new_idx = reversed(new_idx)
```

```python
new_idx
```

```python
df = df.reindex(new_idx)
```

```python
df
```

```python
df.head().iloc[: , :5]
```

```python
df.tail().iloc[:, :5]
```

```python
last_valid = df.apply(pd.Series.last_valid_index)
```

```python
last_valid
```

```python
first_valid = df.apply(pd.Series.first_valid_index)
first_valid
```

```python
last_valid
```

```python
earliest_date = df.index.min()
earliest_date
```

```python
shift_values = last_valid - earliest_date
shift_values
```

```python
print(shift_values)
```

```python
ebola_dict = {}
for idx, col in enumerate(df):
    d = shift_values[idx].days
#     일(days)로 변환한 값을 d에 저장
    shifted = df[col].shift(d)
#     현재 열을 d만큼 이동
    ebola_dict[col] = shifted
```

```python
df_shifted = pd.DataFrame(ebola_dict)
df_shifted
```

```python
df_shifted.tail()
```

```python
df.tail()
```

```python
# 시작일 값이 무조건 최저값으로 변경됨
```

```python
df_shifted
```

```python
df_shifted.index = df_shifted['Day']
df_shifted = df_shifted.drop(['Date', 'Day'], axis=1)
```

```python
df_shifted
```

```python
# 기말고사
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax = df_shifted.iloc[:, :].plot(ax=ax)
ax.legend(fontsize=7, loc=2, borderaxespad=0.)
plt.show()
```

```python
sns.lineplot(data=df_shifted)
```
