```python
df = pd.read_csv('../data/banklist.csv', parse_dates=[5, 6])
df.info()
```

```python
df['closing_quartter'], df['closing_year'] = (df['Closing Date'].dt.quarter, df['Closing Date'].dt.year)
```

```python
df.info()
```

```python
df.sample(10)
```

```python
closing_year = df.groupby(['closing_year']).size()
closing_year
```

```python
import seaborn as sns
sns.histplot(data=df, x='closing_year', y=closing_year['closing_year'])
```

```python
closing_y_p = df.groupby(['closing_year','closing_quartter']).size()
closing_y_p
```

```python
print(closing_y_p)
```

```python
closing_y_p
```

```python
closing_y_p.index
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax = closing_year.plot()
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax = closing_y_p.plot()
plt.show()
```
