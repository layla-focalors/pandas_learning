```python
df = pd.read_csv('../data/weather.csv')
df.head()
```

```python
dfx = df.melt(id_vars=['id','year','month','element'], 
              var_name="day", value_name="temp")
dfx
```

```python
dfu = dfx.pivot_table(index=['id','year', 'month','day'],
                     columns="element",
                     values="temp")
print(dfu)
```

```python
ffx = dfu.reset_index()
```

```python
print(ffx)
```

```python
ffx.columns
```

```python
ffx['mean_temp'] = (ffx['tmax'] + ffx['tmin']) / 2
ffx
```
