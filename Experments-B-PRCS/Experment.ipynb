### 가설 설정
가설 : 데이터의 표준편차와 평균이 같다면, 데이터의 분포가 일치할 것인가?

```python
# 각 데이터별 표준편차와 평균 구하기
import seaborn as sns
import pandas as pd

df = sns.load_dataset("anscombe")
dat1 = df[df['dataset'] == 'I']
dat2 = df[df['dataset'] == 'II']
dat3 = df[df['dataset'] == 'III']
dat4 = df[df['dataset'] == 'IV']

print(f"데이터 1의 표준편차 : {dat1.std()}, 평균 : {dat1.mean()}")
print(f"데이터 2의 표준편차 : {dat2.std()}, 평균 : {dat2.mean()}")
print(f"데이터 3의 표준편차 : {dat3.std()}, 평균 : {dat3.mean()}")
print(f"데이터 4의 표준편차 : {dat4.std()}, 평균 : {dat4.mean()}")
```
