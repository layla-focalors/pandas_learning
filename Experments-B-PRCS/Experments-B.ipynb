```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.lmplot
```

```python
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex', col='size', col_wrap = 2, fit_reg=False)
```

```python
sns.boxplot(x='time', hue='sex', y='tip', data=df)
```

```python
sns.boxplot(x='size', y='tip', hue='sex', data=df)
```

```python
sns.boxplot(x='time', y='tip', hue='sex', data=df)
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', size = df['size'] * 10, data=df)
```
