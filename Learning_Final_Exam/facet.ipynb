```python
import pandas as pd
```

```python
df = pd.read_csv('../data/country_timeseries.csv', parse_dates=[0])
```

```python
df
```

```python
df.index = df['Date']
```

```python
df
```

```python
new_idx = pd.date_range(df.index.min(), df.index.max())
```

```python
new_idx
```

```python
new_idx = reversed(new_idx)
```

```python
df = df.reindex(new_idx)
```

```python
df
```

```python
diff = df.apply(pd.Series.last_valid_index) - df.index.min()
```

```python
df_dict = {}
for i, v in enumerate(df):
    df_dict[v] = df[v].shift(diff[i].days)
```

```python
df_dict
```

```python
df = pd.DataFrame(df_dict)
```

```python
df
```

```python
df.index = df['Day']
```

```python
df
```

```python
df.plot()
```

```python
df = df.drop(['Date', 'Day'], axis=1)
```

```python
df
```

```python
df.plot()
```

```python
fig = df.plot()
fig.legend(fontsize=7, loc=2)
```
