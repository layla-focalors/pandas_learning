```python
def getMean(data):
    sum = 0
    for i in data:
        sum += i
    return sum / len(data)
```

```python
df = pd.read_csv('../data/gapminder.tsv', sep='\t')
```

```python
df.groupby('year').agg(getMean)
```

```python
df.mean()
```

```python
df.groupby('year')['lifeExp'].agg(getMean)
```

```python
df.groupby('year')['lifeExp'].mean()
```

```python
def getMean2(data, n):
    sum = 0
    for i in data:
        sum += i
    return sum / len(data) - n

df.groupby('year')['lifeExp'].agg(getMean2, n=df['lifeExp'].mean())
```

```python
def getMean2(data, n):
    sum = 0
    for i in data:
        sum += i
    return (sum / len(data)) - n

df.groupby('year')['lifeExp'].agg(getMean2, n=df['lifeExp'].mean())
```
