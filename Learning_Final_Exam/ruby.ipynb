```python
df = pd.read_csv('../data/country_timeseries.csv', parse_dates=['Date']) 
df.info() 
df.index = df['Date'] 
new_idx = pd.date_range(df.index.min(), df.index.max()) 
new_idx = reversed(new_idx) 
df = df.reindex(new_idx) 
# 날자 추가 
# tip! reindex 를 해야만 날자가 추가됨!! 
last_valid = df.apply(pd.Series.last_valid_index) 
first_valid = df.apply(pd.Series.first_valid_index) 
earlist_date = df.index.min() 
shift_values = last_valid - df.index.min()
df_dict = {} 
for i, v in enumerate(df): 
    shifted = df[v].shift(shift_values[i].days) 
    df_dict[v] = shifted 
df_dict 
df_n = pd.DataFrame(df_dict) 
df_n.index = df_n['Day'] 
df_n = df_n.drop(['Date', 'Day'], axis=1) 
df_n 
```
