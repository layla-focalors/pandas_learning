```python
df = pd.read_csv('../data/country_timeseries.csv', parse_dates=[0])
```

```python
last_valid = df.apply(pd.Series.last_valid_index)
```

```python
df.index = df['Date']
```

```python
df
```

```python
new_idx = pd.date_range(df.index.min(), df.index.max())
new_idx
new_idx = reversed(new_idx)
```

```python
df = df.reindex(new_idx)
```

```python
df
```

```python
earlist = df.index.min()
shift_value = last_valid - df.index.min()
```

```python
last_valid = df.apply(pd.Series.last_valid_index)
last_valid
```

```python
earlist_date = df.index.min()
shifted_values = last_valid - earlist_date
```

```python
shifted_values
```

```python
df_dict = {}

for i, v in enumerate(df):
    df_dict[v] = df[v].shift(shifted_values[i].days)
```

```python
df_dict
```

```python
df = pd.DataFrame(df_dict)
```

```python
df
```

```python
df.index = df['Day']
```

```python
df = df.drop(['Date', 'Day'], axis=1)
```

```python
df
```

```python
fig = sns.lineplot(data=df)
fig.legend(prop={'size':7})
```
