```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data/country_timeseries.csv', parse_dates=[0])
df.info()
```

```python
df
```

```python
df.index = df['Date']
```

```python
df
```

```python
rann = pd.date_range(df.index.min(), df.index.max())
rann
```

```python
last_valid = df.apply(pd.Series.last_valid_index)
last_valid
```

```python
diff = last_valid - df.index.min()
```

```python
diff
```

```python
df.reindex(reversed(rann))
```

```python
dt_dict = {}
for i, v in enumerate(df):
    dt_dict[v] = df[v].shift(diff[v].days)
```

```python
dt_dict
```

```python
df = pd.DataFrame(dt_dict)
```

```python
df
```

```python
df
```

```python
df.index = df['Day']
```

```python
df
```

```python
df = df.drop(['Date', 'Day'], axis=1)
```
