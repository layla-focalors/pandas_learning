```python
df = pd.read_csv('../data/country_timeseries.csv', parse_dates=[0])
```

```python
df.index = df['Date']
```

```python
df
```

```python
new_idx = pd.date_range(df.index.min(), df.index.max())
```

```python
new_idx
```

```python
diff = df.apply(pd.Series.last_valid_index) - df.index.min()
```

```python
diff
```

```python
df = df.reindex(new_idx)
```

```python
diff
```

```python
df
```

```python
df_main = {}
for i,v in enumerate(df):
    df_main[v] = df[v].shift(diff[i].days)
```

```python
df_main
```

```python
df = pd.DataFrame(df_main)
```

```python
df
```

```python
df.index = df['Day']
```

```python
df
```

```python
df = df.drop(['Date','Day'], axis = 1)
```

```python
df
```

```python
fig = df.plot()
fig.legend(fontsize = 7)
```
