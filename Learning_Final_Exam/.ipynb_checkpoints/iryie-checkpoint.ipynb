```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/country_timeseries.csv', parse_dates=[0])
```

```python
df.info()
```

```python
df.index = df['Date']
```

```python
idx = pd.date_range(df.index.min(), df.index.max())
idx = reversed(idx)
```

```python
idx
```

```python
idx
df = df.reindex(idx)
```

```python
df
```

```python
last_valid = df.apply(pd.Series.last_valid_index)
last_valid
```

```python
total = df.index.min()
```

```python
diff = last_valid - total
```

```python
diff
```

```python
dt_dict = {}
for i,v in enumerate(df):
    dt_dict[v] = df[v].shift(diff[v].days)
```

```python
dt_dict
```

```python
df = pd.DataFrame(dt_dict)
```

```python
df
```

```python
df.index = df['Day']
```

```python
df
```

```python
df = df.drop(['Date', 'Day'], axis=1)
```

```python
fig = df.plot()
fig.legend(fontsize=7, loc=2)
```

```python
fig = sns.lineplot(data=df)
fig.legend(prop={'size':5})
```
