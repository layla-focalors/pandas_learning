```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/country_timeseries.csv', parse_dates=[0])
```

```python
df.info()
```

```python
df.index = df['Date']
idx = pd.date_range(df.index.min(), df.index.max())
idx = reversed(idx)
df.reindex(idx)
last_valid = df.apply(pd.Series.last_valid_index)
```

```python
last_valid
```

```python
diff = last_valid - df.index.min()
```

```python
diff
```

```python
dt_dict = {}

for i, v in enumerate(df):
    dt_dict[v] = df[v].shift(diff[v].days)
dt_dict
```

```python
df = pd.DataFrame(dt_dict)
```

```python
df
```

```python
df = df.drop(['Date', 'Day'], axis=1)
```

```python
df
```

```python
fig = df.plot()
fig.legend(fontsize=7, loc=2)
```

```python
fig = sns.lineplot(data=df)
fig.legend(prop={'size':7})
fig.set_xticklabels(fig.get_xticklabels(), rotation=45)
```
