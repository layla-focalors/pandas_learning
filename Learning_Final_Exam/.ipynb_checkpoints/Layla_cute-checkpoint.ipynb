```python
def getMean(data, n):
    sum = 0
    for i in data:
        sum += i
    return (sum / len(data)) - n
```

```python
df = pd.read_csv('../data/country_timeseries.csv', parse_dates=[0])
```

```python
df
```

```python
df.index = df['Date']
```

```python
df
```

```python
date_range = pd.date_range(df.index.min(), df.index.max())
date_range
```

```python
diff = df.apply(pd.Series.last_valid_index) - df.index.min()
```

```python
diff
```

```python
df = df.reindex(reversed(date_range))
```

```python
df
```

```python
dt_ln = {}
for i, v in enumerate(df):
    dt_ln[v] = df[v].shift(diff[i].days)
dt_ln
```

```python
df = pd.DataFrame(dt_ln)
```

```python
df
```

```python
df.index = df['Day']
```

```python
df = df.drop(['Date','Day'], axis=1)
```

```python
df
```

```python
fig = df.plot()
fig.legend(fontsize=7, loc=2)
```

```python
fig = sns.lineplot(data=df)
fig.legend(prop={'size':7})
```
