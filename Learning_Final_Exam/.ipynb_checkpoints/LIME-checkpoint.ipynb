```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/country_timeseries.csv', parse_dates=[0])
```

```python
df.info()
```

```python
df.index = df['Date']
```

```python
rounder = pd.date_range(df.index.min(), df.index.max())
```

```python
rounder
```

```python
rounder = reversed(rounder)
```

```python
df = df.reindex(rounder)
```

```python
df
```

```python
last_valid = df.apply(pd.Series.last_valid_index)
```

```python
diff = last_valid - df.index.min()
```

```python
dt_imp ={}
for i, v in enumerate(df):
    dt_imp[v] = df[v].shift(diff[i].days)
```

```python
dt_imp
```

```python
df_n = pd.DataFrame(dt_imp)
```

```python
df_n
```

```python
df_n.index = df_n['Day']
```

```python
df_n
```

```python
df_n = df_n.drop(['Date', 'Day'], axis=1)
```

```python
df_n
```

```python
fig = df_n.plot()
fig.legend(fontsize=7, loc=2)
```
