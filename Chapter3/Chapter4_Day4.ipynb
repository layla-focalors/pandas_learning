```python
def sex_color(sex):
    if sex == "Female":
        return 0
    else:
        return 1
```

```python
tips['sex_color'] = tips['sex'].apply(sex_color)
tips.head()
```

```python
scatter = plt.figure()
axes1 = scatter.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'], tips['size'] * 10, tips['sex_color'], alpha=0.5)
axes1.set_title("Total Bill vs Tip Colored by sex and Sized by size")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")
# alpha = 투명도
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title("Scatter of Total Bill and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
```

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg=False)
# 회귀선 제거 : fit_reg
ax.set_title("Scatter of Total Bill and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
```

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel="Total Bill", ylabel='Tip')
joint.fig.subtitle("Joint Plot of Total Bill and Tip", fontsize=10, y=1.03)
```

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips, kind="hex")
joint.set_axis_labels(xlabel="Total Bill", ylabel='Tip')
# kind = hex , 벌집 구조 육각형 그래프(hexbin)
joint.fig.subtitle("Joint Plot of Total Bill and Tip", fontsize=10, y=1.03)
# 엄청 이쁨!!
```
