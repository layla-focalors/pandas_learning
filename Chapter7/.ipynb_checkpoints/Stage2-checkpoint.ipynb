```python
import pandas as pd

# 데이터 로드
# df = pd.read_csv('.

# 'year', 'artist', 'track', 'time', 'date.entered' 열을 기준으로 중복 제거
df = df.drop_duplicates(subset=['year', 'artist', 'track', 'time', 'date.entered'])

# week와 rating 열을 추출하고 정렬
df['week'] = df['week'].str.extract('(\d+)').astype(int)
df = df.sort_values(by=['week', 'rating'])

# 각 열별로 같은 데이터끼리 모아 보기
grouped = df.groupby(['year', 'artist', 'track', 'time', 'date.entered'])
```
