downcast : 정수, 실수와 같은 자료형을 더 작은 형태로 사용
float 64 > float 32 - float64가 더 많은 범위의 실수를 표현할 수 있으나 용량이 크기 때문에 문제 발생

```python
tips_sub_iss['total_bill'] = pd.to_numeric(tips_sub_iss.total_bill, errors='coerce', downcast='float')
```

```python
tips_sub_iss.dtypes
```

```python
tips_sub_iss = pd.to_numeric(tips_sub_iss, errors='coerce', downcast='float')
```
