```python
import seaborn as sns
df = sns.load_dataset("tips")
df
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nx = plt.figure()
nx = sns.pairplot(df)
nx
```

```python
ux = plt.figure()
ux = sns.pairplot(df)
ux.map_upper(sns.kdeplot, fill=True)
ux.map_lower(sns.regplot)
ux.map_diag(sns.distplot)
ux
```
