```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
```

```python
fig = plt.figure()
# x = total_bill , y = tip
axes1 = fig.add_subplot(1, 1, 1)
axes1.plot(tips['total_bill'], tips['tip'], '*', color="pink")
axes1.set_title("x = total_bill, y = tip")
axes1.set_xlabel("total_bill")
axes1.set_ylabel("bill")
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.hist(tips['total_bill'], bins=10, color="pink")
axes1.set_title("Histogram of Total Bill")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Frequency")
fig
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'], color="pink", marker="*")
axes1.set_title("Scatter PLOT!")
axes1.set_xlabel("total bill")
axes1.set_ylabel("Tip")
```

```python
def record_sex(sex):
    if sex == "Female":
        return 0
    else:
        return 1
```
