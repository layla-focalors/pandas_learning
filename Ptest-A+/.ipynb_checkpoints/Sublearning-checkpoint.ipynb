```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.jointplot(x="total_bill", y="tip", data=df, kind="hex", color="pink")
```

```python
sns.kdeplot(x="total_bill", y="tip", data=df, fill=True)
```

```python
sns.barplot(x="time", y="total_bill", data=df, errorbar=None)
```

```python
sns.boxplot(x="time", y="total_bill", data=df, hue='sex')
```

```python
sns.violinplot(x="time", y="total_bill", data=df, split=True, hue='sex')
```

```python
sns.pairplot(df)
```

```python
ax = sns.PairGrid(data=df)
ax.map_upper(sns.regplot)
ax.map_diag(sns.distplot)
ax.map_lower(sns.kdeplot, fill=True)
```
